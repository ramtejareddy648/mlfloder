```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
z=plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("graph b/w X and Y")
plt.show(z)
```

```python
### we add color and style of line and size of of points 

x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y ,color="red",linestyle="--",marker="o",linewidth=4,markersize=9)
## we can use marker symbol as * or anything
```

```python
### we can also apply grid 
plt.plot(x,y,color="orange",linestyle="--",marker="*")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
```

```python
x=[1,3,4,5]
y=[2,3,5,6]
z=[3,4,4,5]
plt.figure(figsize=(9,9))  ##here figure size means width and heigth

plt.subplot(2,2,1)   ###subplot mean how many rows and column and figure position
plt.plot(x,y,color="red")
plt.title("plot 1")


plt.subplot(2,2,2)
plt.plot(x,z ,color="green")
plt.title("plot 2")


plt.subplot(2,2,3)
plt.plot(y,z,color="blue")
plt.title("plot 3")

plt.subplot(2,2,4)
plt.plot(z,x ,color="orange")
plt.title("plot 3")


```

```python
### we can draw bar plots 
x=["A",'B','C','D']
y=[23,12,33,21]

plt.bar(x,y,color="green",edgecolor="black") ## edgecolor means border color of graph
plt.xlabel("products")
plt.ylabel("sales")
plt.title("product and sales")
```

```python
## histogram 


x=[1,1,1,1,2,2,3,4,4,5,5,5]
plt.hist(x,bins=3,color="purple",edgecolor="black")
```

```python
### scatter graph 
#### scatter graph means there is no line only have points

x=[1,2,3,5]
y=[3,5,7,9]

plt.scatter(x,y,color="blue",marker="X")
```

```python
### pie chart

labels=["A","B","C","D"]  ###labels 
size=[12,34,34,45] ### percentage of each label
color=["green","red","orange","purple"]  ####color of each label
exploed=(0.5,0.2,0,0) ###here explod means which label come out from pie chart
plt.pie(size,explode=exploed,colors=color,labels=labels,autopct="%1.2f%%",shadow=True) 
```
