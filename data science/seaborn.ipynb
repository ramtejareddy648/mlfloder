```python
import seaborn as sns

data=sns.load_dataset("tips")
data
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(x="total_bill",y="tip",data=data)
plt.title("total bill and tip")
plt.show()
```

```python
sns.lineplot(x="total_bill",y="tip",data=data)
plt.title("total bill and tip")
plt.show()
```

```python
sns.barplot(x="day",y="total_bill",data=data)
plt.title("total bill and tip")
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=data)
plt.title("total bill and tip")
plt.show()
```

```python
sns>violinplot(x="day",y="total_bill",data=data)
plt.title("total bill and tip")
plt.show()
```

```python
sns.histplot(data["total_bill"],bins=5,kde=True)  ### kde true means lines come if flase no line
```

```python
sns.kdeplot(data["total_bill"],fill=True)
```

```python
sns.kdeplot(data["total_bill"],fill=False)
```
