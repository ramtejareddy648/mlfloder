## isoltion anamaly outlier diction

```python
import pandas as pd
```

```python
df=pd.read_csv('healthcare.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
df.head()
```

```python
from sklearn.ensemble import IsolationForest
forest=IsolationForest(contamination=0.20)
forest.fit(df)
```

```python
predictions=forest.predict(df)
```

```python
predictions
```

```python
import numpy as np
index=np.where(predictions<0)
```

```python
index
```

```python
x=df.values
```

```python
x
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors='r')
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1],c=predictions)
```
