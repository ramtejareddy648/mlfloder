### down sampling

```python
import numpy as np
import pandas as pd

np.random.seed(120)

samples=1000
class_0_ratio=0.9

n1_class_0=int(samples*class_0_ratio)
n2_class_1=samples-n1_class_0

print(f"samplea is {samples}")
print(f'n class 0 ratio is {class_0_ratio}')
print(f'class 0 samples is {n1_class_0}')
print(f"class 1 samoles is {n2_class_1}")
```

```python
class_1=pd.DataFrame({
    'feature1':np.random.normal(loc=1, scale=2,size=n1_class_0),
    "feature2":np.random.normal(loc=0,scale=1,size=n1_class_0),
    "target":0*n1_class_0
})
```

```python
class_1
```

```python
class_2=pd.DataFrame({
    'feature1':np.random.normal(loc=1, scale=2,size=n2_class_1),
    "feature2":np.random.normal(loc=0,scale=1,size=n2_class_1),
    "target":[1]*n2_class_1
})
```

```python
class_2
```

```python
df=pd.concat([class_1,class_2])
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df["target"].value_counts()
```

```python
df[df['target']==0]
```

```python
df[df['target']==1]
```

```python
majority=df[df['target']==0]
minority=df[df['target']==1]
```
