```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
print(data.DESCR)
```

```python
data['feature_names']
```

```python
data['target']
```

```python
data['target_names']
```

```python
df=pd.DataFrame(data=data['data'],columns=data['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
df_scaled=scale.fit_transform(df)
```

```python
df_scaled
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca_data=pca.fit_transform(df_scaled)
```

```python
pca_data
```

```python
import matplotlib.pyplot as plt
plt.scatter(pca_data[:,0],pca_data[:,1],c=data['target'])
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(10,5))
sc.dendrogram(sc.linkage(pca_data,method='ward'))
plt.title('dendrogram')
plt.xlabel('sample')
plt.ylabel('eclidien distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
```

```python
cluster.fit(pca_data)
```

```python
labels=cluster.labels_
```

```python
labels
```

```python
plt.scatter(pca_data[:,0],pca_data[:,1],c=labels)
```

```python
from sklearn.metrics import silhouette_score
feature=[]
for k in range(2,11):
    cluster=AgglomerativeClustering(n_clusters=k,linkage='ward')
    cluster.fit(pca_data)
    score=silhouette_score(pca_data,cluster.labels_)
    feature.append(score)
```

```python
feature
```

```python
plt.plot(range(2,11),feature)
plt.xticks(range(2,11))
```
