### k means cluster

```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=1000,n_features=2,centers=3,random_state=42)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
x_train
```

```python
x_train_scaled=scale.fit_transform(x_train)
```

```python
x_test_scaled=scale.transform(x_test)
```

```python
x_test_scaled
```

```python
x_train_scaled
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel('no of clusters')
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++')
kmeans.fit(x_train_scaled)
```

```python
x_train_pred=kmeans.predict(x_train_scaled)
```

```python
x_train_pred
```

```python
x_test_pred=kmeans.predict(x_test_scaled)
```

```python
x_test_pred
```

```python

from sklearn.metrics import silhouette_score

siljoutee=[]
for k in range(2,11):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train_scaled)
    score=silhouette_score(x_train_scaled,kmeans.labels_)
    siljoutee.append(score)
```

```python
siljoutee
```

```python
plt.plot(range(2,11),siljoutee)
plt.xticks(range(2,11))
plt.xlabel("Number of Cluters")
plt.ylabel("Silhoutte Coeffecient")
plt.show()
```
