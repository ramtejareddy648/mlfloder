```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
one=OneHotEncoder()
one.fit(x[['name','company','fuel_type']])
```

```python
one.categories_
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, StandardScaler

preprocess=ColumnTransformer(transformers=[
    ('oneencoding',OneHotEncoder(categories=one.categories_),['name','company','fuel_type']),
    ('scale',StandardScaler(),['age','kms_driven'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso,Ridge,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
from sklearn.svm import  SVR
```

```python
models={
    'linear':LinearRegression(),
    'lasso':Lasso(),
    'ridge':Ridge(),
    'elasticnet':ElasticNet(),
    'svr':SVR(),
    'decision':DecisionTreeRegressor(),
    'random':RandomForestRegressor(),
    'adboot':AdaBoostRegressor(),
    'graident':GradientBoostingRegressor(),
    'kneigbhor':KNeighborsRegressor()
}
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import r2_score
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    pipe=make_pipeline(preprocess,model)
    pipe.fit(x_train,y_train)
    
    y_train_pred=pipe.predict(x_train)
    y_test_pred=pipe.predict(x_test)
    
    
    print(list(models.keys())[i])
    print('x_train prediction ')
    print('r2score',r2_score(y_train,y_train_pred))
    print('--'*32)
    print('x_test prediction ')
    print('r2_score',r2_score(y_test,y_test_pred))
    print('--end--'*23)
    
    
    
```

```python
from sklearn.ensemble import GradientBoostingRegressor

param_grid = {
    'gradientboostingregressor__n_estimators': [50, 100, 200, 300],
    'gradientboostingregressor__learning_rate': [0.01, 0.1, 0.2, 0.5],
    'gradientboostingregressor__max_depth': [3, 5, 7]
}

```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gradient=GradientBoostingRegressor()
pipe=make_pipeline(preprocess,gradient)
```

```python
grid=GridSearchCV(estimator=pipe,param_grid=param_grid,cv=5)
```

```python
grid.fit(x_train,y_train)
```

```python
y_train_pred=grid.predict(x_train)
y_test_pred=grid.predict(x_test)
```

```python
print('r2_score train data is',r2_score(y_train,y_train_pred))
print('r2score test data is',r2_score(y_test,y_test_pred))
```

```python
param_dist = {
    'decisiontreeregressor__max_depth': [None, 10, 20, 30, 40],
    'decisiontreeregressor__min_samples_split': np.arange(2, 10),
    'decisiontreeregressor__min_samples_leaf': np.arange(1, 10),
    'decisiontreeregressor__max_features': [None, 'sqrt', 'log2']
}
```

```python
decision=DecisionTreeRegressor()
pipe=make_pipeline(preprocess,decision)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_search = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=param_dist,
    n_iter=20, 
    cv=5, 
    scoring='r2', 
    n_jobs=-1, 
    random_state=42
)
```

```python
random_search.fit(x_train,y_train)
```

```python
y_train_pred=random_search.predict(x_train)
y_test_pred=random_search.predict(x_test)
```

```python
print('r2_score train data is',r2_score(y_train,y_train_pred))
print('r2score test data is',r2_score(y_test,y_test_pred))
```
