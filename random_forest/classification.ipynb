```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=45)
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
one=OneHotEncoder(drop='first')
scale=StandardScaler()
```

```python
cat_feature=x.select_dtypes(include='object').columns
num_feature=x.select_dtypes(exclude='object').columns
print(cat_feature,len(cat_feature))
print(num_feature,len(num_feature))
```

```python
process=ColumnTransformer([('onehotencoder',one,cat_feature),
                           ('standardscale',scale,num_feature)],remainder='passthrough')
```

```python
process
```

```python
x_train=process.fit_transform(x_train)
```

```python
x_test=process.transform(x_test)
```

```python
x_train
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
models={
    'logistic':LogisticRegression(),
    'random':RandomForestClassifier(),
    'desicion':DecisionTreeClassifier(),
    'knearest':KNeighborsClassifier(),
    'gussian':GaussianNB()
}
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,f1_score,recall_score
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,y_train)
    
    
    y_train_pred=model.predict(x_train)
    y_test_pred=model.predict(x_test)
    
    print(list(models.keys())[i])
    
    print("prediction for x_train")
    print("classification_report",classification_report(y_train,y_train_pred))
    print('confusion_matrix',confusion_matrix(y_train,y_train_pred))
    print("accuray_score",accuracy_score(y_train,y_train_pred))
    print("precission_score",precision_score(y_train,y_train_pred))
    print("f1_score",f1_score(y_train,y_train_pred))
    print("recall_score",recall_score(y_train,y_train_pred))
    
    print('--'*35)
    
    print('prediction for x_test')
    print("classification_report",classification_report(y_test,y_test_pred))
    print('confusion_matrix',confusion_matrix(y_test,y_test_pred))
    print("accuray_score",accuracy_score(y_test,y_test_pred))
    print("precission_score",precision_score(y_test,y_test_pred))
    print("f1_score",f1_score(y_test,y_test_pred))
    print("recall_score",recall_score(y_test,y_test_pred))
    
    print('=='*35)
    
    
    
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_dist = {
    'n_estimators':randint(50, 500),  # Number of trees
    'max_depth': [None, 10, 20, 30, 50],  # Max depth of each tree
    'min_samples_split': randint(2, 20),  # Minimum samples to split a node
    'min_samples_leaf': randint(1, 10),  # Minimum samples at each leaf node
    'max_features': ['sqrt', 'log2'],  # Number of features to consider at each split
    'bootstrap': [True, False]  # Bootstrap samples when building trees
}
```

```python
param_dist
```

```python
rf=RandomForestClassifier()
rc=RandomizedSearchCV(estimator=rf,param_distributions=param_dist,cv=5,n_iter=100,verbose=3,n_jobs=-1)
```

```python
rc.fit(x_train,y_train)
```

```python
rc.best_params_
```

```python
y_train_pred=rf.predict(x_train)
y_test_pred=rf.predict(x_test)
    

    
print("prediction for x_train")
print("classification_report",classification_report(y_train,y_train_pred))
print('confusion_matrix',confusion_matrix(y_train,y_train_pred))
print("accuray_score",accuracy_score(y_train,y_train_pred))
print("precission_score",precision_score(y_train,y_train_pred))
print("f1_score",f1_score(y_train,y_train_pred))
print("recall_score",recall_score(y_train,y_train_pred))
    
print('--'*35)
    
print('prediction for x_test')
print("classification_report",classification_report(y_test,y_test_pred))
print('confusion_matrix',confusion_matrix(y_test,y_test_pred))
print("accuray_score",accuracy_score(y_test,y_test_pred))
print("precission_score",precision_score(y_test,y_test_pred))
print("f1_score",f1_score(y_test,y_test_pred))
print("recall_score",recall_score(y_test,y_test_pred))
    
print('=='*35)
```
