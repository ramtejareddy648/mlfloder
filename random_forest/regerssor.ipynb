```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=45)
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder,OrdinalEncoder
```

```python
scale=StandardScaler()
Label=LabelEncoder()
ord=OrdinalEncoder()
one=OneHotEncoder(drop='first',handle_unknown='ignore')
```

```python
ord.fit(df[['model']])
```

```python
x=ord.categories_
```

```python
from sklearn.compose import ColumnTransformer

preporcess=ColumnTransformer(transformers=[('standarscale',scale,['vehicle_age','km_driven','mileage','engine','max_power','seats']),
                                           ('ordinal_encode', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1), ['model']),
                                           ('OneHotEncoder',one,['seller_type','fuel_type','transmission_type'])],remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
models={
    'linear':LinearRegression(),
    'tree':DecisionTreeRegressor(),
    'random':RandomForestRegressor(),
    'knn':KNeighborsRegressor()
    
}
```

```python

from sklearn.pipeline import make_pipeline
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    pipeline=make_pipeline(preporcess,model)
    pipeline.fit(x_train,y_train)
    
    y_train_pred=pipeline.predict(x_train)
    y_test_pred=pipeline.predict(x_test)
    
    print('model is',list(models.keys())[i])
    print('x_train prediction')
    print('r2_score',r2_score(y_train,y_train_pred))
    print('mean_absoulute_error',mean_absolute_error(y_train,y_train_pred))
    print('mean_squared_error',mean_squared_error(y_train,y_train_pred))
    
    print('-'*35)
    
    print('x_test prdiction')
    print('model is',list(models.keys())[i])

    print('r2_score',r2_score(y_test,y_test_pred))
    print('mean_absoulute_error',mean_absolute_error(y_test,y_test_pred))
    print('mean_squared_error',mean_squared_error(y_test,y_test_pred))
    print("=="*35)
    
    
```

```python
param_grid = {
    'n_estimators': [50, 100, 200, 300],  # Number of trees
    'max_depth': [10, 20, 30, None],  # Depth of trees
    'min_samples_split': [2, 5, 10],  # Minimum samples to split a node
    'min_samples_leaf': [1, 2, 4],  # Minimum samples at a leaf node
    'max_features': ['sqrt', 'log2', None],  # Number of features to consider at each split
    'bootstrap': [True, False]  # Whether to bootstrap samples
}
```

```python
random=RandomForestRegressor()
from sklearn.model_selection import RandomizedSearchCV
rc=RandomizedSearchCV(estimator=random,param_distributions=param_grid,cv=5,n_iter=100,n_jobs=-1)

```

```python
pipeline=make_pipeline(preporcess,rc)
```

```python
pipeline.fit(x_train,y_train)
    
y_train_pred=pipeline.predict(x_train)
y_test_pred=pipeline.predict(x_test)
    
##print('model is',list(models.keys())[i])
print('x_train prediction')
print('r2_score',r2_score(y_train,y_train_pred))
print('mean_absoulute_error',mean_absolute_error(y_train,y_train_pred))
print('mean_squared_error',mean_squared_error(y_train,y_train_pred))
    
print('-'*35)
    
print('x_test prdiction')
##print('model is',list(models.keys())[i])

print('r2_score',r2_score(y_test,y_test_pred))
print('mean_absoulute_error',mean_absolute_error(y_test,y_test_pred))
print('mean_squared_error',mean_squared_error(y_test,y_test_pred))
print("=="*35)
```
